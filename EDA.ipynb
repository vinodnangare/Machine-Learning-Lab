```python
import pandas as pd
import matplotlib.pyplot as plt

# Group by 'type' and count the number of shows in each category
grouped_type = df.groupby('type')['show_id'].count()

# Plotting the result as a bar chart
grouped_type.plot(kind='bar', color='skyblue')
plt.title("Number of Shows by Type")
plt.xlabel("Type")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

```python
# Group by 'type' and 'country' and count the number of shows in each category
grouped_type_country = df.groupby(['type', 'country'])['show_id'].count().unstack()

# Plotting a stacked bar chart
grouped_type_country.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title("Number of Shows by Type and Country")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
# Group by 'release_year' and count the number of shows for each year
grouped_year = df.groupby('release_year')['show_id'].count().sort_index(ascending=False)

# Plotting the result as a line chart
grouped_year.plot(kind='line', marker='o', color='orange')
plt.title("Number of Shows by Year of Release")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
# Group by 'rating' and count the number of shows
grouped_rating = df.groupby('rating')['show_id'].count().sort_values(ascending=False)

# Plotting the result as a horizontal bar chart
grouped_rating.plot(kind='barh', color='lightcoral')
plt.title("Number of Shows by Rating")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.show()
```

```python
# Group by 'listed_in' (genre) and count the number of shows
grouped_genre = df.groupby('listed_in')['show_id'].count()

# Plotting the result as a pie chart
grouped_genre.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title("Genre Distribution")
plt.ylabel("")  # Remove y-axis label
plt.show()
```

```python
# Group by 'director' and count the number of shows for each director
grouped_director = df.groupby('director')['show_id'].count().sort_values(ascending=False)

# Plotting the top 10 directors with most shows
grouped_director.head(10).plot(kind='bar', color='purple')
plt.title("Top 10 Directors by Number of Shows")
plt.xlabel("Director")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
